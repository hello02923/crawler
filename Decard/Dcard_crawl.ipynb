```python
post_id = '235200000'
result_df = pd.DataFrame()
for i in range(40):
    print(post_id,end=',')
    req = requests.get('https://www.dcard.tw/_api/forums/alcohol/posts?before='+post_id+'&popular=false')
    data = req.text
    df = pd.read_json(data)
    result_df = pd.concat([result_df, df],sort=False)
    if len(df) != 0:
        num = df['id'][-1:].values[0]
        post_id = f'{num}'
    else:
        break
```

```python
result_df = result_df.reset_index(drop=True)
```

```python
result_df
```

```python
result_df.to_csv('dcard_alcohol.csv',encoding = 'utf-8-sig')
```
